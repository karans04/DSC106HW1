```python
import pandas as pd
```

```python
rows = []
for i,f in enumerate(open('ucsd_cds_admission_data_2005_to_2018','r')):
    if i == 0:
        cols = f.replace('\n','').split(',')
    else:
        rows.append(f.split(','))
        
```

```python
pd.DataFrame(rows, columns = cols)
```
